## Historical Index Line Charts (Interactive)

```python
# Import Altair
import altair as alt
alt.renderers.enable('notebook')
```

### Price-to-income Index

```python
# Massage the data
pti_us = data_pti.loc[data_pti.RegionName == 'United States']
data_pti = data_pti.loc[data_pti.RegionName != 'United States']
data_pti.reset_index(drop=True, inplace=True)
```

```python
# Indices
pti_len = data_pti.shape[0]
pti_max = 0
pti_25 = int(pti_len * 0.25)
pti_median = int(pti_len * 0.5)
pti_75 = int(pti_len * 0.75)
pti_min = pti_len - 1
```

```python
pti_ts = pti_us.append(data_pti.loc[[pti_max, pti_25, pti_median, pti_75, pti_min]])

pti_ts.reset_index(drop=True, inplace=True)
pti_ts.drop(columns=['RegionID', 'SizeRank', 'Index', 'HistoricAverage_1985thru1999'], inplace=True)

pti_ts_t = pti_ts.transpose()
pti_ts_t.columns = pti_ts_t.iloc[0]
pti_ts_t = pti_ts_t.iloc[1:]

pti_ts_t.reset_index(inplace=True)

pti_ts_t.rename(index=str, columns={'index': 'Time'}, inplace=True)
```

```python
pti_long = pti_ts_t.melt('Time', var_name='Area', value_name='Price-to-income Index')

pti_long['Price-to-income Index'] = pti_long['Price-to-income Index'].astype(float).round(2)

pti_long.to_csv('pti_ts_data.csv', index=False)
```

#### Generate Interactive Plot

```python
# Create a selection that chooses the nearest point & selects based on x-value
nearest = alt.selection(type='single', nearest=True, on='mouseover',
                        fields=['Time'], empty='none')

# The basic line
line = alt.Chart(pti_long).mark_line(interpolate='basis').encode(
    x='Time:T',
    y='Price-to-income Index:Q',
    color='Area:N'
)

# Transparent selectors across the chart. This is what tells us
# the x-value of the cursor
selectors = alt.Chart(pti_long).mark_point().encode(
    x='Time:T',
    opacity=alt.value(0),
).add_selection(
    nearest
)

# Draw points on the line, and highlight based on selection
points = line.mark_point().encode(
    opacity=alt.condition(nearest, alt.value(1), alt.value(0))
)

# Draw text labels near the points, and highlight based on selection
text = line.mark_text(align='left', dx=5, dy=-5).encode(
    text=alt.condition(nearest, 'Price-to-income Index:Q', alt.value(' '))
)

# Draw a rule at the location of the selection
rules = alt.Chart(pti_long).mark_rule(color='gray').encode(
    x='Time:T',
).transform_filter(
    nearest
)

# Put the five layers into a chart and bind the data
alt.layer(
    line, selectors, points, rules, text
).properties(
    width=600, height=400
)
```

## Correlation Matrix

```python
data_pti_values = data_pti.drop(columns=['RegionName',
                                         'SizeRank',
                                         'Index',
                                         'HistoricAverage_1985thru1999',])
```

```python
data_ma_values = data_ma.drop(columns=['RegionName',
                                         'SizeRank',
                                         'Index',
                                         'HistoricAverage_1985thru1999',])
```

```python
data_ra_values = data_ra.drop(columns=['RegionName',
                                         'SizeRank',
                                         'Index',
                                         'HistoricAverage_1985thru1999',])
```

```python
# Find the common RegionID (intersection) among PTI, MA and RA
common_region_ids = np.intersect1d(data_pti_values.RegionID.values,
                                   np.intersect1d(data_ma_values.RegionID.values, data_ra_values.RegionID.values))
```

```python
# PTI
data_pti_values.index = data_pti_values.RegionID
data_pti_values.drop(columns=['RegionID'], inplace=True)
data_pti_values = data_pti_values.loc[common_region_ids]
data_pti_values.sort_index(inplace=True)

data_pti_values = data_pti_values.values.ravel()
```

```python
# MA
data_ma_values.index = data_ma_values.RegionID
data_ma_values.drop(columns=['RegionID'], inplace=True)
data_ma_values = data_ma_values.loc[common_region_ids]
data_ma_values.sort_index(inplace=True)

data_ma_values = data_ma_values.values.ravel()
```

```python
# RA
data_ra_values.index = data_ra_values.RegionID
data_ra_values.drop(columns=['RegionID'], inplace=True)
data_ra_values = data_ra_values.loc[common_region_ids]
data_ra_values.sort_index(inplace=True)

data_ra_values = data_ra_values.values.ravel()
```

```python
# Construct a DataFrame for correlation matrix
corr_df = pd.DataFrame({'Price-to-income Index': data_pti_values,
                        'Mortgage Affordability': data_ma_values,
                        'Renter Affordability': data_ra_values})
```

```python
corr_df.dropna(how='any', inplace=True)
```

```python
# Plot a correlation heatmap
corr_matrix = corr_df.corr()
```

```python
sns.set()
ax = sns.heatmap(corr_matrix, linewidths=.5, cmap='Blues', annot=True, fmt='.4g', vmin=0)
ax.set_title('Correlation Heatmap')
```
